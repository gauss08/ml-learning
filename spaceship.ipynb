```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score,StratifiedKFold
from sklearn.ensemble import RandomForestClassifier,VotingClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC,LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

import warnings
#warnings.filterwarnings('ignore')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
log=LogisticRegression(max_iter=1000,penalty='l2',C=1)
log.fit(x_train,y_train)
log_p=log.predict(x_test)
log_ac=accuracy_score(log_p,y_test)
print(log_ac)
```

```python
knn=KNeighborsClassifier()
knn_p={
    'n_neighbors':range(1,15)
}
clf_knn=GridSearchCV(knn,knn_p)
clf_knn.fit(x_train,y_train)
best_knn=clf_knn.best_estimator_
```

```python
best_knn.fit(x_train,y_train)
knn_p=best_knn.predict(x_test)
knn_ac=accuracy_score(knn_p,y_test)
print(knn_ac)
```

```python
l_svc=LinearSVC()
params={'penalty':['l2','l1'],
       'C':[0.001,0.01,0.1,1,10,100],
        'dual':[True,False],
       'max_iter':[1000,1500,2000]}
clf_l_svc=GridSearchCV(l_svc,params)
clf_l_svc.fit(x_train,y_train)

best_l_svc=clf_l_svc.best_estimator_
```

```python
best_l_svc.fit(x_train,y_train)
l_svc_p=best_l_svc.predict(x_test)
l_svc_ac=accuracy_score(l_svc_p,y_test)
print(l_svc_ac)
```

```python
svc=SVC()
params={
    'C':[0.1,1,10,100,200],
    'probability':[True],
    'gamma':[0.1,1,10,100,200]
}

clf_svc=GridSearchCV(svc,params,n_jobs=6)
clf_svc.fit(x_train,y_train)

best_svc=clf_svc.best_estimator_
```

```python
best_svc.fit(x_train,y_train)
svc_p=best_svc.predict(x_test)
svc_ac=accuracy_score(svc_p,y_test)
print(svc_ac)
```

```python
from sklearn.gaussian_process.kernels import RBF
kernel = 1.0 * RBF(1.0)

gpc=GaussianProcessClassifier(kernel=kernel)
gpc.fit(x_train,y_train)
gpc_p=gpc.predict(x_test)
gpc_ac=accuracy_score(gpc_p,y_test)
print(gpc_ac)
```
